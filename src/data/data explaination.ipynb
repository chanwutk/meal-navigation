## Recipe

Note there are "n/a" for units. For instace, 1 onion. Because there is not any units, so we use n/a to represent.

Also, note the unit for milk in recipe is cup, but the unit of milk in the ingredient is bottle. there are some ingredient has different units, so I add a feature "addable".

Also, note that there is a feature "addable", which is to prevent the situation that we need 1 cups of milk and 1 cups of milk, but we need to buy 2 bottles of milk. For all ingredient that only consumed a little in one meal, I set the addable to false.

To be more clear, 2 addable product + 3 addable product = 5 addable products.
But, 2 unadaable product + 3 unaddable product = 1 unaddable product.

Also, the boolean value in constraints is in the order of:
1. Vegetarian
2. Cilantro
3. Lamb
4. Pork

Where True means it could offer to both True/False people, false means it could only offer to the constraint people(for instance, true for vegetarian and false for pork).
